```python
import datetime
import json
import matplotlib.pyplot as plt
```

```python
with open('../tracked_partitions.json', 'r') as f:
    tracked_partitions = json.load(f)

print(f'Loaded {len(tracked_partitions)} tracked partitions')
```

```python
lengths = [len(p[1]) for p in tracked_partitions.items()]
print(f'Lengths: {lengths}')
```

```python
filtered = [p for p in tracked_partitions.items() if len(p[1]) > 1]
print(f'Filtered: {len(filtered)}')
```

```python
def plot_partition(partition):
    partition_data = partition[1]
    times = [datetime.datetime.fromisoformat(d[0]) for d in partition_data]
    values = [d[1]["period"]["value"] for d in partition_data]
    plt.plot(times, values)
```

```python
for partition in filtered:
    #print([p[0] for p in partition[1]])
    plot_partition(partition)

plt.xlim([datetime.datetime.fromisoformat('2022-09-22T00:00:00'), datetime.datetime.fromisoformat('2022-09-24T00:00:00')])
plt.show()
```
